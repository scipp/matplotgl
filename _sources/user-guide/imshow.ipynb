# Imshow

```python
import matplotgl.pyplot as plt
import numpy as np
```

## Basic imshow

```python
fig, ax = plt.subplots()

a = np.random.random((200, 300))

ax.imshow(a, extent=[10, 50, 200, 500])

fig
```

## With a colorbar

```python
fig, ax = plt.subplots()

im = ax.imshow(np.random.random((20, 20)))
fig.colorbar(im)

fig
```
